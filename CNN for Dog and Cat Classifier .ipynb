### This model classifies cats from dogs, using the Kaggle dataset of 25,000 cat and dog images.

#### The approach I used was a multi layer CNN leveraging Keras for the model architecture, as well as the image feed and preprocessing advantages.

#### The model is then tested on a couple of new images, including a doggo that is close to my heart.

```python
### Keras packages for deep learning and classification

from keras.preprocessing.image import ImageDataGenerator
from keras.models import Sequential
from keras.layers import Conv2D, MaxPooling2D
from keras.layers import Activation, Dropout, Flatten, Dense
from keras import backend as K
```

```python
### Check if Keras is using GPU
K.tensorflow_backend._get_available_gpus()
```

#### Global parameters

```python
train_data_dir = 'data/train'
validation_data_dir = 'data/validation'
epochs = 50
batch_size = 16
img_width, img_height = 150, 150
nb_train_samples = 2000
nb_validation_samples = 500
```

#### Build the functions needed to shape and normalise the data to be passed into the neural network

#### We are going to use the Keras function to augment, and feed the data into the model

```python
train_data = ImageDataGenerator(
    rescale=1. / 255,
    shear_range=0.2,
    zoom_range=0.2,
    horizontal_flip=True)

test_data = ImageDataGenerator(rescale=1. / 255)
```

```python
if K.image_data_format() == 'channels_first':
    input_shape = (3, img_width, img_height)
else:
    input_shape = (img_width, img_height, 3)
```

#### Define the Model

```python
model = Sequential()
model.add(Conv2D(32, (3, 3), input_shape=input_shape))
model.add(Activation('relu'))
model.add(MaxPooling2D(pool_size=(2, 2)))

model.add(Conv2D(32, (3, 3)))
model.add(Activation('relu'))
model.add(MaxPooling2D(pool_size=(2, 2)))

model.add(Conv2D(64, (3, 3)))
model.add(Activation('relu'))
model.add(MaxPooling2D(pool_size=(2, 2)))

model.add(Conv2D(32, (3, 3)))
model.add(Activation('relu'))
model.add(MaxPooling2D(pool_size=(2, 2)))

model.add(Conv2D(64, (3, 3)))
model.add(Activation('relu'))
model.add(MaxPooling2D(pool_size=(2, 2)))

model.add(Flatten())
model.add(Dense(64))
model.add(Activation('relu'))
model.add(Dropout(0.5))
model.add(Dense(1))
model.add(Activation('sigmoid'))
```

```python
model.compile(loss='binary_crossentropy',
              optimizer='adam',
               metrics=['accuracy'])
```

#### Show summary of model architecture

```python
model.summary()
```

#### Define data feed, classes, and then train model

```python
train_generator = train_data.flow_from_directory(
    train_data_dir,
    target_size=(img_width, img_height),
    batch_size=batch_size,
    class_mode='binary')

validation_generator = test_data.flow_from_directory(
    validation_data_dir,
    target_size=(img_width, img_height),
    batch_size=batch_size,
    class_mode='binary')

print(train_generator.class_indices)
```

```python
model.fit_generator(
    train_generator,
    steps_per_epoch=nb_train_samples // batch_size,
    epochs=epochs,
    validation_data=validation_generator,
    validation_steps=nb_validation_samples // batch_size,
    verbose = 1)
```

#### Got a 92.80% accuracy with the model, however as I do with all my models need to actual test it in a real world setting by looking at images it has never seen downloaded from both Google images and also from my own collection

```python
from IPython.display import Image
import numpy as np
from keras.preprocessing import image
```

#### Define a function which we can test images easily

```python
def prediction(image_directory):
    img = image.load_img(image_directory, target_size=(img_width,img_height))
    image_rescale = image.img_to_array(img)
    image_rescale = np.expand_dims(image_rescale,axis=0)
    image_test = np.vstack([image_rescale])
    prediction = model.predict_classes(image_test)
    if prediction == 1:
        print("This is a doggo, as I predicted a {}".format(prediction))
    elif prediction ==0:
        print("This is a cat, as I predicted a {}".format(prediction))
    
```
